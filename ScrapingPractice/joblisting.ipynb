```python
import requests
from bs4 import BeautifulSoup

url = 'https://realpython.github.io/fake-jobs/'

response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')
card_elements = soup.find_all('div', class_='column is-half')
job_listings = []
    
for card in card_elements:
    position = card.find('h2', class_='title is-5').text
    name = card.find('h3', class_='subtitle is-6 company').text
    location = card.find('p', class_='location').text.strip()
    date_posted = card.find('time')['datetime']
    image_url = card.find('img')['src']
        
    temp_data = [name, position, location, date_posted, image_url]
    job_listings.append(temp_data)

for i in job_listings:
    print(temp_data)
```
